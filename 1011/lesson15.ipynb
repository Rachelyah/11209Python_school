```python
#用open()讀寫檔案，最少需要一個引數，其他都有預設值可以不用寫
'''
open(file, mode='r', buffering=- 1, encoding=None, errors=None, newline=None, closefd=True, opener=None)
'''

#newlineNone=''＞新的一行最好寫成空字串
#後面的呼叫，可以不一定要照順序，但你就要寫引述名稱的呼叫


try:
    file=open('students.csv', mode='r', encoding='utf-8', newline='')
    file.close() #實體方法
    print(file.closed)   #實體屬性＞檢查是不是已經關閉，會是True or False
except:
    print('讀檔錯誤')

#讀完資料一定要關閉->資料名稱.close()->實體方法method
#可以用->資料名稱.close->實體屬性，檢查是否為True->已關閉

#可以寫try：如果讀檔錯誤，這個檔案就不會被建立
#這樣可以避免妳之後寫app時整個應用程式crush，出錯時會出現錯誤訊息
```

```python
#檔案物件的實體：read() 讀取並傳出檔案中所有的字串

try:
    file=open('students.csv', mode='r', encoding='utf-8', newline='')
    print(file.read()) #只能讀一次，讀完就不能再讀
    file.close() #關閉了就不能讀取
    print(file.closed)   #實體屬性＞檢查是不是已經關閉，會是True or False
except:
    print('讀檔錯誤')
```

```python
#檔案物件的實體：readline() 自動讀取下一行，可以把每行分開

try:
    file=open('students.csv', mode='r', encoding='utf-8', newline='')
    print(file.readline())
    print(file.readline())
    print(file.readline())
    file.close() #關閉了就不能讀取
    print(file.closed)   #實體屬性＞檢查是不是已經關閉，會是True or False
except:
    print('讀檔錯誤')
```

```python
#csv module的使用
#很可惜字串不是python的資料結構
#csv.reader(csvfile, dialect='excel', **fmtparams)
#iterate(可迭代的)代表可以用for in迴圈抓裡面的值

import csv
try:
    file=open('students.csv', mode='r', encoding='utf-8', newline='')
    csv_reader = csv.reader(file)
    for row in csv_reader:
        print(row)
    file.close()
    print(file.closed)
except:
    print('讀檔錯誤')
```

#### 如上範例，csv_reader儲存二維的資料


```python
#csv module的使用
#很可惜字串不是python的資料結構
#csv.reader(csvfile, dialect='excel', **fmtparams)
#iterate(可迭代的)代表可以用for in迴圈抓裡面的值

import csv
try:
    file=open('students.csv', mode='r', encoding='utf-8', newline='')
    csv_reader = csv.reader(file)
    for row in csv_reader:
        print(row)
    file.close()
    print(file.closed)
except:
    print('讀檔錯誤')
```

```python
#很可惜，我的資料型別都是字串，不是python儲存資料的方式
#所以我要把這些字串，儲存成list的形式

import csv
try:
    file=open('students.csv', mode='r', encoding='utf-8', newline='')
    csv_reader = csv.reader(file)
    csv_list=list(csv_reader)  #建立csv_list，把csv_reader的內容儲存成listy資料型別
    file.close()
    print(file.closed)
except:
    print('讀檔錯誤')

print(csv_list)
```

```python
#csv.DictReader()
#把指定csv資料轉為dict字典形式，與list的差別是他有前面的項目：
#class csv.DictReader(f, fieldnames=None, restkey=None, restval=None, dialect='excel', *args, **kwds)


import csv
try:
    file=open('students.csv', mode='r', encoding='utf-8', newline='')
    csv_dictReader = csv.DictReader(file)
    csv_list=list(csv_dictReader)           #把dict的資料再用list儲存
    file.close()
except:
    print('讀檔錯誤')

print(csv_list)
```
